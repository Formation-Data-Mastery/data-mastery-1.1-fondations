## Étape 5 — Python et uv

Python est le langage de programmation utilisé pour l'analyse de données. `uv` est un gestionnaire de projets Python moderne qui remplace `pip` et `venv`.

### Pourquoi uv plutôt que pip ?

| | pip + venv (méthode classique) | uv (méthode recommandée) |
|--|-------------------------------|---------------------------|
| **Créer un projet** | `python -m venv .venv` puis `source .venv/bin/activate` (2 commandes) | `uv init mon-projet` (1 commande) |
| **Installer une bibliothèque** | `pip install pandas` (30-60 secondes) | `uv add pandas` (2-5 secondes) |
| **Fichier de dépendances** | `requirements.txt` à maintenir manuellement | `pyproject.toml` géré automatiquement |
| **Conflits de versions** | Fréquents, difficiles à résoudre | Résolution automatique |

### Installation de uv

**Mac / Linux** — exécuter dans le terminal :
```bash
curl -LsSf https://astral.sh/uv/install.sh | sh
```

**Windows** (PowerShell) :
```powershell
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

Après l'installation, **fermer et rouvrir le terminal** pour que la commande `uv` soit disponible.

````python
# Vérification que uv est installé
import subprocess

try:
    result = subprocess.run(["uv", "--version"], capture_output=True, text=True)
    print("uv installé — version :", result.stdout.strip())
except FileNotFoundError:
    print("uv non trouvé — fermer et rouvrir le terminal après l'installation")
````

## Étape 6 — Créer le premier projet

Un projet Python regroupe le code, les données et les dépendances dans un dossier isolé. C'est la bonne pratique pour éviter les conflits entre projets.

Exécuter ces commandes dans le terminal :
```bash
mkdir -p ~/projets
cd ~/projets
uv init mon-projet-data
cd mon-projet-data
uv add pandas plotly ipykernel
```

Ces commandes :
- Créent le dossier `~/projets/mon-projet-data/`
- Initialisent un projet Python avec ses fichiers de configuration (`pyproject.toml`)
- Installent **pandas** (manipulation de données), **plotly** (graphiques interactifs) et **ipykernel** (permet à VS Code d'exécuter les notebooks)

### Comprendre l'environnement virtuel (.venv)

Quand `uv init` crée un projet, il génère un dossier `.venv/` à l'intérieur. C'est un **environnement virtuel** : une copie isolée de Python avec ses propres bibliothèques.

**Pourquoi c'est important :**
- Le projet A peut utiliser pandas 2.0, le projet B pandas 1.5 — sans conflit
- Les dépendances du projet sont documentées dans `pyproject.toml` — quelqu'un d'autre peut recréer le même environnement
- Si quelque chose casse, on peut supprimer `.venv/` et le recréer avec `uv sync`

**En pratique :** avec `uv`, l'environnement virtuel est géré automatiquement. La commande `uv run` exécute toujours le code dans le bon environnement, sans avoir besoin de l'activer manuellement.

### Structure du projet après initialisation

```
mon-projet-data/
├── .venv/              ← environnement virtuel (géré par uv, ne pas modifier)
├── pyproject.toml      ← liste des dépendances du projet
├── uv.lock             ← versions exactes installées (généré automatiquement)
└── hello.py            ← le code source
```

### Git dans VS Code

VS Code intègre Git nativement. Dès qu'un projet contient un dossier `.git/`, le panneau **Source Control** (icône de branches dans la barre latérale gauche) affiche les fichiers modifiés.

**Ce que fait Git :** c'est un système de versionnage. Il enregistre l'historique de toutes les modifications du projet. Chaque enregistrement s'appelle un **commit** — c'est un point de sauvegarde auquel on peut revenir.

**Ce qu'on voit dans VS Code :**
- Les fichiers **modifiés** apparaissent avec un **M** (Modified)
- Les fichiers **nouveaux** apparaissent avec un **U** (Untracked)
- On peut cliquer sur un fichier pour voir ce qui a changé (diff)

**Bonne nouvelle :** `uv init` initialise Git automatiquement. Le projet contient déjà un dossier `.git/` et un fichier `.gitignore`. Il suffit de faire le premier commit :

### Script de vérification automatique

Ce script vérifie que tous les composants sont correctement installés. Exécuter la cellule suivante pour un diagnostic complet.

````python
# Script de vérification complète de l'environnement Data Mastery
# Exécuter cette cellule pour diagnostiquer l'installation

import subprocess
import sys
import importlib

VERT  = "\033[92m"
ROUGE = "\033[91m"
JAUNE = "\033[93m"
RESET = "\033[0m"

def verifier_commande(nom, commande):
    """Vérifie qu'une commande est accessible dans le terminal."""
    try:
        result = subprocess.run(commande, capture_output=True, text=True)
        version = result.stdout.strip().split("\n")[0]
        print(f"{VERT}[OK]{RESET} {nom} : {version}")
        return True
    except FileNotFoundError:
        print(f"{ROUGE}[MANQUANT]{RESET} {nom} : non installé")
        return False

def verifier_module_python(nom_module):
    """Vérifie qu'un module Python est importable."""
    try:
        module = importlib.import_module(nom_module)
        version = getattr(module, "__version__", "version inconnue")
        print(f"{VERT}[OK]{RESET} Python module '{nom_module}' : {version}")
        return True
    except ImportError:
        print(f"{ROUGE}[MANQUANT]{RESET} Python module '{nom_module}' : non installé")
        print(f"        -> Installer avec : uv add {nom_module}")
        return False

print("=" * 55)
print("  VÉRIFICATION DE L'ENVIRONNEMENT DATA MASTERY")
print("=" * 55)
print()

print("--- Commandes système ---")
ok_node   = verifier_commande("Node.js",     ["node",   "--version"])
ok_claude = verifier_commande("Claude Code", ["claude", "--version"])
ok_uv     = verifier_commande("uv",          ["uv",     "--version"])

print()
print("--- Python ---")
print(f"{VERT}[OK]{RESET} Python : {sys.version.split()[0]}")
version = sys.version_info
if version.major == 3 and version.minor >= 12:
    print(f"{VERT}[OK]{RESET} Version Python >= 3.12")
else:
    print(f"{JAUNE}[AVERTISSEMENT]{RESET} Python {version.major}.{version.minor} — 3.12+ recommandé")

print()
print("--- Bibliothèques Python ---")
ok_pandas = verifier_module_python("pandas")
ok_plotly = verifier_module_python("plotly")

print()
print("=" * 55)
tous_ok = ok_node and ok_claude and ok_uv and ok_pandas and ok_plotly
if tous_ok:
    print(f"{VERT}Environnement complet — prêt pour la formation.{RESET}")
else:
    print(f"{JAUNE}Des composants sont manquants — suivre les étapes ci-dessus.{RESET}")
print("=" * 55)
````
